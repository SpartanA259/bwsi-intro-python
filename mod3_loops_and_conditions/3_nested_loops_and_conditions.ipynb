## FizzBuzz Problem

Write a short program that prints each number from 1 to 100 except:

- For each multiple of 3, print "Fizz" instead of the number

- For each multiple of 5, print "Buzz" instead of the number

- For numbers which are multiples of both 3 and 5, print "FizzBuzz" instead of the number

### Q.1 Please outline in words the steps that should be executed for each iteration of the loop

### A.1 For each loop iteration, my code should check first if it's divisible by 3 and 5 and print FizzBuzz if true, check if it's divisible by 3 and print Fizz if true, check if it's divisible by 5 and print Buzz if true, and lastly print the number

### Q.2 Please implement the solution described in Q.1 with comments

```python
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0: # check if divisible by both 3 and 5
        print("FizzBuzz")           # print FizzBuzz and go to next iteration
    elif i % 3 == 0:                # check if divisible by 3
        print("Fizz")               # print Fizz and go to next iteration
    elif i % 5 == 0:                # check if divisible by 5
        print("Buzz")               # print Buzz and go to next iteration
    else:
        print(i)                    # print number
```

### Q.3 Please complete the commented lines below with descriptions of the code as indicated. 

```python
# Create list to store prime numbers
prime_numbers_list = []

# This for loop iterates over the range from 2 to 10
for n in range(2,11):
    # Set is_prime as True
    is_prime = True
    # This for loop iterates over the range from 2 to the parent loop iteration number
    for k in range(2, n):
        # This condition checks if the initial number is divisible by each number before it starting at 2
        if n % k == 0:
            # If n % k == 0, then print k times a number that equal k and complete the parent loop
            print(f"{n} = {k} * {n // k}")
            is_prime = False
            break
    # This condition checks if is_prime is True
    if is_prime:
        # If n is prime, then print as output 
        # and append to list of prime numbers
        print(f"{n} is a prime number")
        prime_numbers_list.append(n)
        
print(f"\nList of prime numbers in range: {prime_numbers_list}")
```
